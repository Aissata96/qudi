```python
def aligner_plot_matrix(voltage_matrix,scan_matrix):
    voltage_list=voltage_matrix.reshape(voltage_matrix.shape[0]*voltage_matrix.shape[1])
    scan_list=voltage_matrix.reshape(scan_matrix.shape[0]*scan_matrix.shape[1])
    binwidth=min(abs(voltage_list[0:-1]-voltage_list[1::]))
    bins=np.linspace(min(voltage_list),max(voltage_list),
                     int((min(voltage_list)-max(voltage_list))/binwidth)+1)
    index=np.digitize(voltage_list)
    hist=np.zeros(len(bins))
    hist[index-1]+=scan_list
    return[hist]
    
```
